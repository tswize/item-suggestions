```python
import pandas as pd
import json
from collections import defaultdict
import random
```

```python
workstep_df = pd.read_csv("worksteps_10k.csv", header=None)

# TO DO: consider how to access the worksteps given the workbooks (using SPy library) 
```

```python
def getWorkstep(json_val):
    return json.loads(json.loads(json_val)['stringValue'])
    #json.loads(json.loads(workstep_df.iloc[0,0])['stringValue'])
```

```python
workstep_df = workstep_df.rename(columns={0: "json"})
workstep_df["dict"] = workstep_df["json"].apply(getWorkstep) 
workstep_df
```

```python
# detailed view of one workstep dictionary
wkstp_dict = workstep_df.iloc[4, 1]
wkstp_dict
```

```python
def get_id_dict(wrkstp_dict): 
    id_dict = defaultdict(list)
    if 'state' in wrkstp_dict.keys():
        keys_dict = wrkstp_dict['state']['stores'] #keys_dict is the nested dict containing the keys (sqTrendSeriesScore, etc ...)
        for key in keys_dict.keys():
            if 'items' in list(keys_dict[key]):
                items = keys_dict[key]['items']
                for item in items:
                    id_dict[key].append(item['id'])
    return dict(id_dict)
```

```python
# test method on detailed dictionary above
get_id_dict(wkstp_dict)
```

```python
workstep_df["ids"] = workstep_df["dict"].apply(get_id_dict)
workstep_df
```
